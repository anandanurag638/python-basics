```python
import pandas as pd

def clean_data(df):
    # Identify outliers in the 'streams' column using the IQR method
    Q1 = df['streams'].quantile(0.25)
    Q3 = df['streams'].quantile(0.75)
    IQR = Q3 - Q1
    # Filter rows that are outliers
    outliers = df[(df['streams'] < (Q1 - 1.5 * IQR)) | (df['streams'] > (Q3 + 1.5 * IQR))]
    # Assign the result back to df
    df = outliers
    return df

# Loaded variable 'df' from URI: c:\Users\anura\Downloads\archive (6)\spotify_global_trends_2026 - spotify_global_trends_final.csv
df = pd.read_csv(r'c:\Users\anura\Downloads\archive (6)\spotify_global_trends_2026 - spotify_global_trends_final.csv')

df_clean = clean_data(df.copy())
df_clean.head()
df['artist_name'].value_counts().head(5)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. डेटा लोड करें (मान लीजिए df आपका DataFrame है)
# अगर Genre कॉलम में लिस्ट है, तो उसे साफ करना होगा
# मान लेते हैं कॉलम का नाम 'genre' है

# 2. Genre के आधार पर कुल Streams का हिसाब (Total Popularity)
genre_popularity = df.groupby('genre')['streams'].sum().sort_values(ascending=False).head(10)

print("Top 10 Most Popular Genres by Total Streams:")
print(genre_popularity)

# 3. Visualizing Genre Popularity
plt.figure(figsize=(12,6))
sns.barplot(x=genre_popularity.values, y=genre_popularity.index, palette='viridis')
plt.title('Top 10 Genres on Spotify (2026)')
plt.xlabel('Total Streams')
plt.ylabel('Genre')
plt.show()
import matplotlib.pyplot as plt
import seaborn as sns

# मान लीजिए genre_popularity आपकी टॉप 10 लिस्ट है
plt.figure(figsize=(12,6))
sns.barplot(x=genre_popularity.head(10).values, y=genre_popularity.head(10).index, palette='magma')
plt.title('Top 10 Music Genres - Global Trends 2026')
plt.xlabel('Total Streams')
plt.show()
avg_viral = df.groupby('genre')['viral_score'].mean().sort_values(ascending=False).head(10)

print("Top 10 Genres by Average Viral Score:")
print(avg_viral)
print("Top 10 Genres by Average Viral Score:")
print(avg_viral)
print(avg_viral)
plt.figure(figsize=(12,6))
sns.barplot(x=avg_viral.values, y=avg_viral.index, palette='viridis')# Viral Score vs Total Streams का मुकाबला
comparison = df.groupby('genre').agg({'viral_score': 'mean', 'streams': 'sum'}).sort_values(by='viral_score', ascending=False).head(10)
print(comparison)
plt.title('Top 10 Most Viral Genres on Spotify (2026)')
plt.xlabel('Average Viral Score')
plt.ylabel('Genre')
plt.show()# Viral Score vs Total Streams का मुकाबला
comparison = df.groupby('genre').agg({'viral_score': 'mean', 'streams': 'sum'}).sort_values(by='viral_score', ascending=False).head(10)
print(comparison)
# Correlation निकालना (0 से 1 के बीच)
correlation = df['viral_score'].corr(df['streams'])
print(f"Correlation between Viral Score and Streams: {correlation}")

# इसे विजुअलाइज़ करने के लिए Scatter Plot
plt.figure(figsize=(10,6))
sns.scatterplot(x='viral_score', y='streams', data=df)
plt.title('Relationship: Viral Score vs Total Streams')
plt.show()
import matplotlib.pyplot as plt
import seaborn as sns

# 1. डेटा को Millions में बदलें (ताकि स्केल समझ आए)
df['streams_m'] = df['streams'] / 1_000_000

# 2. सबसे ज़्यादा स्ट्रीम्स वाला गाना (Outlier) ढूँढें

#
# 3. उस अकेले डॉट पर नाम लिखें (Annotation)
#op_song['track_name']} by {top_song['artist_name']}")
import plotly.express as px

fig = px.scatter(df, 
                 x="viral_score", 
                 y="streams", 
                 hover_name="track_name", # यह हेडलाइन में दिखेगा
                 hover_data=["artist_name", "streams"], # अतिरिक्त जानकारी
                 title="Spotify 2026: Click on dots to see Song Names",
                 labels={"viral_score": "Viral Score (0-5)", "streams": "Total Streams"},
                 template="plotly_white")

# 2. डॉट्स का रंग और साइज थोड़ा बेहतर करें
fig.update_traces(marker=dict(size=10, color='royalblue', line=dict(width=1, color='DarkSlateGrey')))

# 3. चार्ट दिखाएं
fig.show()# 1. 'The Superstars' (High Viral Score + High Streams)
superstars = df[(df['viral_score'] > 3) & (df['streams'] > df['streams'].median())]

# 2. 'The Hidden Gems' (Low Viral Score but High Streams - Long term hits)
hidden_gems = df[(df['viral_score'] < 1.5) & (df['streams'] > df['streams'].median())]

print("--- Superstars ---")
print(superstars[['track_name', 'artist_name']].head())

print("\n--- Hidden Gems ---")
print(hidden_gems[['track_name', 'artist_name']].head())

fig = px.scatter(df, x="viral_score", y="streams", 
                 trendline="ols",
                 hover_name="track_name",
                 title="Predicting Success: Viral Score vs Streams")
fig.show()
# High Viral Score (>3) लेकिन Low Streams (Median से कम)
one_hit_wonders = df[(df['viral_score'] > 3) & (df['streams'] < df['streams'].median())]

print("--- One Hit Wonders (High Hype, Low Value) ---")
print(one_hit_wonders[['track_name', 'artist_name']].head())
%pip install python-pptx
from pptx import Presentation
```
